```python
Z1 = complex(10, 5) 
Z2 = complex(4, -3) 
Z3 = complex(6, 2)
choice = input("Enter connection type (series/parallel): ")
if choice == "series":
    Z_total = Z1 + Z2 + Z3
    print("Total Impedance (Series) =", Z_total)
elif choice == "parallel":
    Z_total = 1 / (1/Z1 + 1/Z2 + 1/Z3)
    print("Total Impedance (Parallel) =", Z_total)
else:
    print("Invalid choice")
```
