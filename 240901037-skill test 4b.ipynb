```python
import pandas as pd
data = {'Product': ['Laptop','Smartphone','Tablet','Monitor','Keyboard'],
'Price': [1200,800,300,150,50],
'Quantity': [10,20,15,25,50]
       }
df = pd.DataFrame(data)
print("Inventory Data")
print(df)
# Operation 1: Average price
print("Average Price:", df['Price'].mean())
# Operation 2: Maximum price
print("Maximum Price:", df['Price'].max())
# Operation 3: Total quantity
print("Total Quantity:", df['Quantity'].sum())
# Operation 4: Products with price greater than 300
print("Products with price > 300")
print(df[df['Price'] > 300])
# Operation 5: Add total value column
df['Total_Value'] = df['Price'] * df['Quantity']
print("Inventory with Total Value")
print(df)
# Operation 6: Total inventory value
print("Total Inventory Value:", df['Total_Value'].sum())
```
